#  Analyzing Total sales

```python
# total_sales is computed
total_sales = data['Sales'].sum()

# Creating a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=['Total Sales'], y=[total_sales], palette='Blues_d')

# Adding title and labels
plt.title('Total Sales')
plt.ylabel('Sales ($)')
plt.xlabel('')

# Annotating the total sales value
plt.text(0, total_sales/2, f'${total_sales:,.2f}', ha='center', color='white', fontsize=14)

plt.show()
```

# Total Sales Yearly

```python
# Assuming your data is in a DataFrame named 'data'
data = pd.DataFrame({
    'Order Date': pd.date_range(start='2011-01-01', end='2014-12-31', freq='D'),
    'Sales': np.random.randint(100, 1000, size=1461)  # Random sales data for example
})

# Ensure 'Order Date' is in datetime format
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Grouping by year
yearly_sales = data.resample('Y', on='Order Date')['Sales'].sum()

# Plotting yearly sales
plt.figure(figsize=(12, 6))
sns.barplot(x=yearly_sales.index.year, y=yearly_sales.values, palette='Blues_d')

# Adding title and labels
plt.title('Yearly Total Sales')
plt.ylabel('Sales ($)')
plt.xlabel('Year')

# Annotating each bar with the sales value
for i in range(len(yearly_sales)):
    plt.text(i, yearly_sales[i], f'${yearly_sales[i]:,.2f}', ha='center', va='bottom', fontsize=10)

plt.show()
```

# Analyzing Patterns and Sales trend over time

```python
# Converting date columns to datetime
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%m/%d/%Y')
data['Ship Date'] = pd.to_datetime(data['Ship Date'], format='%m/%d/%Y')

# Extract year and month for trend analysis
data['Order Year'] = data['Order Date'].dt.year
data['Order Month'] = data['Order Date'].dt.month

# Calculating total sales and profit by year
annual_sales = data.groupby('Order Year')['Sales'].sum().reset_index()
annual_profit = data.groupby('Order Year')['Profit'].sum().reset_index()

# Plotting sales and profit over time
plt.figure(figsize=(14, 7))

```

```python
# Sales over time
plt.subplot(1, 2, 1)
sns.lineplot(data=annual_sales, x='Order Year', y='Sales', marker='o')
plt.title('Annual Sales')
plt.xlabel('Year')
plt.ylabel('Total Sales')

# Profit over time
plt.subplot(1, 2, 2)
sns.lineplot(data=annual_profit, x='Order Year', y='Profit', marker='o', color='orange')
plt.title('Annual Profit')
plt.xlabel('Year')
plt.ylabel('Total Profit')

plt.tight_layout()
plt.show()
```

```python
# Monthly Sales trend
monthly_sales = data.groupby(['Order Year', 'Order Month'])['Sales'].sum().reset_index()
monthly_sales_pivot = monthly_sales.pivot(index='Order Month', columns='Order Year', values='Sales')

plt.figure(figsize=(14, 7))
sns.heatmap(monthly_sales_pivot, cmap='YlGnBu', annot=True, fmt=".1f")
plt.title('Monthly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

```python
# Category-wise Sales Analysis
category_sales = data.groupby('Category')['Sales'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=category_sales, x='Category', y='Sales')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Region-wise Sales Analysis
region_sales = data.groupby('Region')['Sales'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=region_sales, x='Region', y='Sales')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Country-wise Sales Analysis
# Sorting the country_sales DataFrame by sales value
country_sales_sorted = country_sales.sort_values(by='Sales', ascending=False)

# Selecting the top 10 countries
top_10_countries = country_sales_sorted.head(10)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_countries, x='Country', y='Sales')
plt.title('Top 10 Countries by Total Sales')
plt.xlabel('Country')
plt.ylabel('Total Sales')
# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
