# Impact of Income and Overtime on Attrition

```python
# Set a low income threshold (e.g., 25th percentile of Monthly_Income)
low_income_threshold = data['Monthly_Income'].quantile(0.25)

# Create a new column for identifying employees who work overtime and have low income
data['Overworked_Underpaid'] = (data['Over_Time'] == 'Yes') & (data['Monthly_Income'] < low_income_threshold)

# Plotting the scatter plot with hue for Attrition and style for Overworked_Underpaid
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data, x='Monthly_Income', y='Years_At_Company', hue='Attrition', style='Overworked_Underpaid', palette='Set1', s=100, alpha=0.6)

# Add regression lines
sns.regplot(data=data[data['Overworked_Underpaid'] == True], x='Monthly_Income', y='Years_At_Company', scatter=False, color='red', label='Overworked & Underpaid')
sns.regplot(data=data[data['Overworked_Underpaid'] == False], x='Monthly_Income', y='Years_At_Company', scatter=False, color='blue', label='Others')

# Add plot labels and title
plt.xlabel('Monthly Income')
plt.ylabel('Years at Company')
plt.title('Impact of Income and Overtime on Attrition')
plt.legend(title='Attrition & Overworked/Underpaid')
plt.grid(True)

# Show the plot
plt.show()
```
